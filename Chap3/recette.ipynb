Soit $x$ le prix de hausse.
<ol>
    <li> Exprimer en fonction de $x$ le prix de la place.</li>
    <li> Exprimer en fonction de $x$ le nombre de spectateurs</li>
    <li> En déduire la recette en fonction de $x$.</li>
    <li> Démontrer que la recette en fonction du prix $p$ est $$R(p)=-3p^2+210p.$$</li>
    <li> Pour quelle valeur de $p$, la recette sera-t-elle nulle ?</li>
</ol>

```python
X = [k / 100 for k in range(1000, 6001)]
Y = [-3 * p**2 + 210 * p for p in X]
plt.plot(X, Y, color = 'blue')
```
